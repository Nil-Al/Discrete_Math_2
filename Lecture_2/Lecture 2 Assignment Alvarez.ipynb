Consider the following experiment:

1. The experiment consists of $n$ identical trials.
2. The outcome of each trial falls into one of $k$ classes or cells.
3. The probability that the outcome of a single trial falls into cell $i$, is $p_i,i = 1,2,...,k$ and remains the same from trial to trial. Notice that $p_1+p_2+p_3+···+p_k =1.$
4. The trials are independent.
5. The random variables of interest are $Y_1 , Y_2 , . . . , Y_k$, where $Y_i$ equals the number of trials for which the outcome falls into cell $i$. Notice that $Y_1 +Y_2 +Y_3 +···+Y_k =n$.
6. The joint probability distribution $P(y) = \frac{n!}{y_1!y_2!y_3!...y_n!} P_1^{y_1} P_2^{y_2} P_3^{y_3}....P_n^{y_n}$

Question 1: Show that the expected value $E[y] = \Sigma y \cdotp P(y) = n \cdot p_i$ <br> <br>
To show that the expected value $E[Y_i] = \sum_y y_i P(y) = n \cdot p_i$, we need to use the given joint probability distribution and the definition of expected value.
The joint probability distribution is given as:
$$P(y) = \frac{n!}{y_1!y_2!y_3!...y_k!} p_1^{y_1} p_2^{y_2} p_3^{y_3}....p_k^{y_k}$$
where $y = (y_1, y_2, \ldots, y_k)$ represents the number of trials in each cell, and $\sum_{i=1}^k y_i = n$.
The expected value of $Y_i$ is defined as:
$$E[Y_i] = \sum_y y_i P(y)$$
We can expand this sum by considering all possible values of $y_1, y_2, \ldots, y_k$ such that $\sum_{i=1}^k y_i = n$:
$$E[Y_i] = \sum_{y_1=0}^n \sum_{y_2=0}^{n-y_1} \sum_{y_3=0}^{n-y_1-y_2} \ldots \sum_{y_k=n-\sum_{j=1}^{k-1} y_j} y_i \frac{n!}{y_1!y_2!y_3!...y_k!} p_1^{y_1} p_2^{y_2} p_3^{y_3}....p_k^{y_k}$$
Now, let's focus on the term involving $y_i$:
$$y_i \frac{n!}{y_1!y_2!y_3!...y_k!} p_1^{y_1} p_2^{y_2} p_3^{y_3}....p_k^{y_k}$$
We can factor out $p_i^{y_i}$ and $y_i!$ from this term:
$$y_i \frac{n!}{y_1!y_2!y_3!...y_k!} p_1^{y_1} p_2^{y_2} p_3^{y_3}....p_k^{y_k} = \frac{n!}{(y_i-1)!(y_1!y_2!y_3!...y_{i-1}!y_{i+1}!...y_k!)} p_1^{y_1} p_2^{y_2} p_3^{y_3}....p_i^{y_i-1}....p_k^{y_k}$$
Now, we can recognize the term on the right-hand side as the joint probability distribution for $y_1, y_2, \ldots, y_{i-1}, y_i-1, y_{i+1}, \ldots, y_k$, which we can denote as $P(y_1, y_2, \ldots, y_{i-1}, y_i-1, y_{i+1}, \ldots, y_k)$.
Therefore, we can rewrite the expected value as:
$$E[Y_i] = \sum_{y_1=0}^n \sum_{y_2=0}^{n-y_1} \sum_{y_3=0}^{n-y_1-y_2} \ldots \sum_{y_k=n-\sum_{j=1}^{k-1} y_j} (y_i-1+1) P(y_1, y_2, \ldots, y_{i-1}, y_i-1, y_{i+1}, \ldots, y_k) p_i$$

2. Each trial results in one of two outcomes: success, $S$, or failure, $F$.
3. The probability of success on a single trial is equal to some value $p$ and remains the same from trial to trial. The probability of a failure is equal to q = $(1 − p)$.
4. The trials are independent.
5. The random variable of interest is $Y$ , the number of successes observed during the $n$ trials.

Question 1: From the following steps shown above, derive the probability distribution of the experiment. <br> <br>

Question 2: Show that the expectation of this probability distribution is $E[y] = \Sigma y \cdotp P(y) = np$. <br> <br>

To show that the expectation of the binomial probability distribution is $E[Y] = \sum_y y \cdot P(y) = np$, we can use the following steps:
The probability mass function (PMF) of the binomial distribution is given by:
$$P(Y = y) = \binom{n}{y} p^y q^{n-y}$$
where $q = 1 - p$.
The expectation of a discrete random variable $Y$ is defined as:
$$E[Y] = \sum_y y \cdot P(y)$$
Substituting the binomial PMF, we get:
$$E[Y] = \sum_{y=0}^{n} y \cdot \binom{n}{y} p^y q^{n-y}$$
We can differentiate the binomial expansion $(p + q)^n$ with respect to $p$:
$$(p + q)^n = \sum_{y=0}^{n} \binom{n}{y} p^y q^{n-y}$$
Differentiating both sides with respect to $p$, we get:
$$n(p + q)^{n-1} = \sum_{y=0}^{n} y \cdot \binom{n}{y} p^{y-1} q^{n-y}$$
Multiplying both sides by $p$, we have:
$$np(p + q)^{n-1} = \sum_{y=0}^{n} y \cdot \binom{n}{y} p^y q^{n-y}$$
Since $p + q = 1$, we can simplify the right-hand side:
$$np(p + q)^{n-1} = np^n = \sum_{y=0}^{n} y \cdot \binom{n}{y} p^y q^{n-y}$$
Therefore, we have shown that:
$$E[Y] = \sum_{y=0}^{n} y \cdot P(y) = np$$
Hence, the expectation of the binomial probability distribution is $np$, where $n$ is the number of trials, and $p$ is the probability of success in a single trial.

Question 3: Experience has shown that 30% of all persons afflicted by a certain illness recover. A drug company has developed a new medication. Ten people with the illness were selected at random and received the medication; nine recovered shortly thereafter. Suppose that the medication was absolutely worthless. What is the probability that at least nine of ten receiving the medication will recover? <br> <br>

Let $X$ be the random variable representing the number of recoveries among 10 people, with the probability of recovery being 0.3 (without the medication).
We want to find $P(X \geq 9)$, the probability of observing at least 9 recoveries out of 10 people.
$$P(X \geq 9) = P(X = 9) + P(X = 10)$$
$$P(X \geq 9) = \binom{10}{9} (0.3)^9 (0.7)^{1} + \binom{10}{10} (0.3)^{10}$$
$$P(X \geq 9) = 120 (0.3)^9 (0.7)^{1} + (0.3)^{10} \approx 0.0014$$
Therefore, $P(X \geq 9) \approx 0.0014$ or 0.14%.

Question 4: Suppose that a lot of 5000 electrical fuses contains 5% defectives. If a sample of 5 fuses is tested, find the probability of observing at least one defective. <br> <br>

Therefore, $P(X \geq 1) \approx 0.2262$ or 22.62%.

Task 3: Consider a probility distribution of $p(y) = \frac{\lambda^y}{y!} e^{-\lambda}$.

Question 1: Find a general formula for the expected value of this distribution. <br> <br>

Step 1: $E[Y] = \sum_{y=0}^{\infty} y \cdot p(y)$
Step 2: Substitute $p(y) = \frac{\lambda^y}{y!} e^{-\lambda}$
$$E[Y] = \sum_{y=0}^{\infty} y \cdot \frac{\lambda^y}{y!} e^{-\lambda}$$
Step 3: Use the property $\lambda e^{\lambda} = \frac{d}{d\lambda} (\lambda e^{\lambda}) - e^{\lambda}$
$$E[Y] = \sum_{y=0}^{\infty} y \cdot \frac{1}{y!} \left(\frac{d}{d\lambda} (\lambda e^{\lambda}) - e^{\lambda}\right)$$
Step 4: Simplify by taking the derivative term out of the summation
$$E[Y] = \frac{d}{d\lambda} (\lambda e^{\lambda}) - \sum_{y=0}^{\infty} \frac{\lambda^y}{y!} e^{-\lambda}$$
Step 5: Recognize $\sum_{y=0}^{\infty} \frac{\lambda^y}{y!} e^{-\lambda} = e^{\lambda}$
$$E[Y] = \frac{d}{d\lambda} (\lambda e^{\lambda}) - e^{\lambda}$$
Step 6: Apply the property again
$$E[Y] = \lambda$$
Therefore, the expected value of the probability distribution $p(y) = \frac{\lambda^y}{y!} e^{-\lambda}$ is $E[Y] = \lambda$.
