A graph in Graph Theory is denoted as $G = (V,E)$ where $V$ represents as the vertices/nodes and $E$ represents as the edges or connection/link.

```python
import matplotlib.pyplot as plt
import networkx as nx
n = 5  # 10 nodes
m = 1  # 20 edges
seed =55  # seed random number generators for reproducibility

# Use seed for reproducibility
G = nx.gnm_random_graph(n, m, seed=seed)
# Add self-loops to the remaining nodes
edgelist = [(1, 2), (3, 3), (2, 4), (0, 0)]
G.add_edges_from(edgelist)
pos = nx.spring_layout(G, seed=seed)  # Seed for reproducible layout
nx.draw(G, pos=pos,with_labels=True)
plt.show()
```

$\textbf{Directed Graph (Digraphs)}$
-

A directed graph is a graph but with directions. An edge from $1$ to $0$ is denoted as $(0,1)$ which indicates that there is a connection from $0$ to $1$.

```python
import matplotlib.pyplot as plt
import networkx as nx
seed = 55
# Use seed for reproducibility
G = nx.DiGraph()

# Add nodes
G.add_nodes_from([0,1,2])
# Add self-loops to the remaining nodes
edgelist = [(0, 1),(1,2),(2,0)]
G.add_edges_from(edgelist)
#pos = nx.spring_layout(G, seed=seed)  # Seed for reproducible layout
#pos = nx.planar_layout(G)
pos = nx.circular_layout(G)
nx.draw(G, pos=pos,with_labels=True)
plt.show()
```

- $e_1$ = $(0, 1)$
- $e_2$ = $(1, 2)$
- $e_3$ = $(2, 0)$

$\textbf{Simple Graphs}$

Formula to get how many simple graphs you can draw: <br>
$\begin{pmatrix}vertices \\ edges \end{pmatrix} = \frac{vertices!}{(edges!)(vertices-edges)!}$

$\textbf{Complete Graphs}$
-

- A complete graph on $n$ vertices, is denoted as $K_n$.
- A simple graph that has $n$ vertices and exactly one edge connecting each pair of vertices that are distinct.

```python
import matplotlib.pyplot as plt
import networkx as nx
seed = 55
# Use seed for reproducibility
G = nx.Graph()

# Add nodes
G.add_nodes_from([1,2,3,4,5])
G.add_node (0)
# Add self-loops to the remaining nodes
edgelist = [(1,2),(3,4),(3,5),(4,5)]
G.add_edges_from(edgelist)
pos = nx.planar_layout(G)  # Seed for reproducible layout
nx.draw(G, pos=pos,with_labels=True)
plt.show()
```

$\textbf{Bipartite Graphs}$
-

- Set of vertices can be divided into two disjoint sets such that no two vertices within the same set are adjacent.
- Denoted as $K_n,_m$.



```python
import matplotlib.pyplot as plt
import networkx as nx

kabit_network = nx.Graph()
kabit_network.add_nodes_from(['A','B'], bipartite=0) # Add the node attribute "bipartite"
kabit_network.add_nodes_from(['X','Y','Z'], bipartite=1)
kabit_network.add_edges_from([('A','X'), ('A','Y'), ('A','Z')
                              , ('B','X'), ('B','Y'), ('B','Z')])

top = nx.bipartite.sets(kabit_network)[0]
pos = nx.bipartite_layout(kabit_network, top)
nx.draw(kabit_network, pos=pos, with_labels=True, node_color=['pink','magenta','red','deepskyblue','cyan'])
plt.show()
```

$\textbf{The Handshake Theorem}$

If $G$ is any graph, then the sum of the degrees of all the vertices of $G$ equals twice the number of edges of $G$. Specifically, if the vertices of $G$ are $v_1,v_2,...,v_n$, where $n$ is a nonnegative integer, then
the total degree of $G = deg(v_1) + deg(v_2 ) +    + deg(v_n )$ = 2$\cdotp$(the number of edges of $G$).

The Handshake theorem can be used to determine if a Graph exists because the total degree should always be even.

$Example$

1. A graph with five vertices of degrees 2, 4, 2, 1, and 1 exists because the total degree is 10 (1+1+2+2+4)

2. A graph with 10 vertices of degrees 1, 1, 3, 2, 3, 1, 2, 2, 3, and 3 does not exist because the total degree is 21 (1+1+1+2+2+2+3+3+3)


$\textbf{Travelling in a Graph}$
-


- A $\textbf{walk}$ in a graph is a sequence of vertices connected by edges.
- A $\textbf{trail}$ is a walk with distinct edges.
- A $\textbf{path}$ is a trail with distinct vertices.
- A $\textbf{closed walk}$ is a walk that starts and ends at the same vertex.
- A $\textbf{circuit}$ is a closed walk with distinct edges (except for the start/end vertex).
- A $\textbf{simple circuit}$ is a circuit with distinct vertices (except for the start/end vertex), and no repeated edges.

$\textbf{Euler's Circuit}$
-

- A circuit that traverses every $\textbf{edge}$ of a graph exactly once, starting and ending at the same vertex.

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a new graph
G = nx.Graph()
# Add nodes
for node in ['A', 'B', 'C']:
    G.add_node(node)

# Add edges between nodes
edges = [('A', 'B'), ('B', 'C'), ('C', 'A')]

# Assuming all nodes are interconnected as per the complex relationships
for edge in edges:
    G.add_edge(*edge)

# Draw the network
nx.draw(G,with_labels=True, node_color='lightblue', edge_color='gray')
plt.show()
```

$\textbf{Solution:}$ <br>
$\textbf{A,B,C,A}$

$\textbf{Hamiltonian Circuits}$
-

- A circuit that visits every $\textbf{vertex}$ of a graph exactly once, starting and ending at the same vertex.

```python
import matplotlib.pyplot as plt
import networkx as nx

seed = 55
# Use seed for reproducibility
G = nx.Graph()

# Add nodes
G.add_nodes_from(["A", "B", "C", "D", "E"])  # Adding one more node
# Add self-loops to the remaining nodes
edgelist = [("A", "B"), ("A", "C"), ("A", "D"), ("B", "C"), ("B", "D"), ("C", "D"), ("D", "E"), ("A", "E")]  # Adding one more edge
G.add_edges_from(edgelist)
pos = nx.spring_layout(G)  # Change layout to spring layout

# Draw the graph with updated layout
nx.draw(
    G, pos, edge_color='black', width=1, linewidths=1,
    node_size=500, node_color='pink', alpha=0.9,
    with_labels=True
)

plt.show()
```

Route:
$A \to B \to C \to D \to E \to A$
